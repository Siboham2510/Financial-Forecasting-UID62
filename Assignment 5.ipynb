```python
import math
from scipy.stats import norm
```

```python
class BlackScholes:
    def __init__(self, S0, K, T, r, sigma):
        self.S0 = S0  # Current stock price
        self.K = K    # Strike price
        self.T = T    # Time to expiration in years
        self.r = r    # Risk-free interest rate
        self.sigma = sigma  # Volatility

    def d1(self):
        return (math.log(self.S0 / self.K) + (self.r + 0.5 * self.sigma**2) * self.T) / (self.sigma * math.sqrt(self.T))

    def d2(self):
        return self.d1() - self.sigma * math.sqrt(self.T)

    def call_price(self):
        return self.S0 * norm.cdf(self.d1()) - self.K * math.exp(-self.r * self.T) * norm.cdf(self.d2())

    def put_price(self):
        return self.K * math.exp(-self.r * self.T) * norm.cdf(-self.d2()) - self.S0 * norm.cdf(-self.d1())
```

```python
from datetime import datetime

# Given dates
valuation_date = datetime(2023, 2, 1)  # 1st Feb 2023
expiry_date = datetime(2023, 2, 23)   # 23rd Feb 2023

# Calculate time to expiration in years
days_to_expiration = (expiry_date - valuation_date).days
T = days_to_expiration / 365
```

```python
# Given parameters
S0 = 24141.95  # Underlying spot price
K = 25000      # Example strike price
r = 0.07       # Risk-free interest rate
sigma = 0.15   # Volatility

# Initialize Black-Scholes model
bs = BlackScholes(S0, K, T, r, sigma)
```

```python
# Calculate option prices
call_price = bs.call_price()
put_price = bs.put_price()

print(f"Call Option Price: {call_price:.2f}")
print(f"Put Option Price: {put_price:.2f}")
```
