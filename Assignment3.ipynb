```python
# Clean column names by stripping spaces
data.columns = data.columns.str.strip()

# Convert 'Expiry' to numeric, coercing errors to NaN
data['Expiry'] = pd.to_datetime(data['Expiry'], format='%d-%b-%Y', errors='coerce')

# Convert 'Date' column to datetime
data['Date'] = pd.to_datetime(data['Date'], format='%d-%b-%Y', errors='coerce')
```

```python
# Function to calculate the model price using the no-arbitrage principle for futures
def calculate_model_price(spot_price, expiry, date, interest_rate):
    # Calculate the time to maturity in days
    time_to_maturity = (expiry - date).days / 365  # Difference in days converted to years

    # Ensure time_to_maturity is positive, if not return NaN
    if time_to_maturity < 0:
        return np.nan

    # Calculate the model price using the no-arbitrage formula
    return spot_price * np.exp(interest_rate * time_to_maturity)

# Calculate Model Price and other metrics
data["Model Price"] = data.apply(
    lambda row: calculate_model_price(row["Underlying Value"], row["Expiry"],row["Date"], INTEREST_RATE), axis=1
)
# Keep the existing Settle Price intact and just calculate Model Price
data["Spot Price"] = data["Underlying Value"]
```

```python
# Calculate time to maturity in years (difference between Expiry and Date)
data['Time to Maturity'] = (data['Expiry'] - data['Date']).dt.days / 365

# Calculate the Derived Interest Rate based on the formula
data['Derived Interest Rate'] = np.log(data['Settle Price'] / data['Spot Price']) / data['Time to Maturity']

# Calculate the mean of the derived interest rates
mean_interest_rate = abs(data['Derived Interest Rate'].mean())

# Given interest rate
given_interest_rate = 4.40 / 100  # 4.40% converted to decimal

# Calculate the percentage error
percentage_error = abs(mean_interest_rate - given_interest_rate) / given_interest_rate * 100

# Display the results
print(f"Mean Derived Interest Rate: {mean_interest_rate:.6f}")
print(f"Percentage Error: {percentage_error:.2f}%")
```
